### 匯入套件

```python
from lxml import etree
import requests
import json
import pandas as pd
```

### 運算過程

```python
def get_all_api():
    alljson=[]
    for page in range(1,totalpage()):
        apilink="https://fund.api.cnyes.com/fund/api/v2/search/fund?order=priceDate&sort=desc&page="+str(page)+"&institutional=0&isShowTag=1&fields=categoryAbbr,change,changePercent,classCurrencyLocal,cnyesId,displayNameLocal,displayNameLocalWithKwd,forSale,forSale,inceptionDate,investmentArea,lastUpdate,nav,prevPrice,priceDate,return1Month"
        res = requests.get(apilink)
        resjson=json.loads(res.text)
        alljson.append(resjson['items']['data'])
    return alljson
def totalpage():
    link = 'https://fund.cnyes.com/search/'
    res = requests.get(link)
    raw_data = etree.HTML(res.content.decode('utf-8'))
    totalpage="".join(raw_data.xpath("//strong[@data-reactid='579']/text()"))
    if int(totalpage)%20==0:
        totalpage=int(totalpage)/20
    else:
        totalpage=(int(totalpage)//20)+1
    return totalpage+1
def fitdf():
    dfa = pd.DataFrame.from_dict(get_all_api()[0])
    for pagejson in range(1,len(get_all_api())):
        dfb = pd.DataFrame.from_dict(get_all_api()[pagejson])
        dfa = pd.concat([dfa, dfb], axis= 0)
    dfa = dfa.reset_index(drop=True)
    return dfa
```

### to csv

```python
df = pd.DataFrame(fitdf())
df.to_csv('fund.csv',encoding = 'utf-8')
```
